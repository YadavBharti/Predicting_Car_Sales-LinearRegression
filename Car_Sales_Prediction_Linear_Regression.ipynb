```python
# Handling Outliers

def outlier_capping(x):
        x = x.clip_upper(x.quantile(0.99))
        x = x.clip_lower(x.quantile(0.01))
        return x

    
car_sales_num=car_sales_num.apply(lambda x : outlier_capping(x))    
 
```

```python
car_sales_num.describe()
```

```python
# Handling missing values

def Missing_Imputation(x):
    x = x.fillna(x.mean())
    return x
    
car_sales_num = car_sales_num.apply(lambda x : Missing_Imputation(x))     
```

```python
car_sales_num.describe()
```

```python
pandas_profiling.ProfileReport(car_sales_num)
```

```python
pandas_profiling.ProfileReport(car_sales_cat)
```

```python
# Handling Categorical Variables
def create_dummies(fm, colname):
    dummies=pd.get_dummies(fm[colname], prefix=colname)
    dummies.drop(dummies.columns[0], axis=1, inplace=True)
    fm= pd.concat([fm, dummies], axis=1)
    fm.drop(colname, axis=1, inplace=True)
    return fm

car_sales_cat=car_sales[['Manufacturer', 'Vehicle_type']]
for columns in ['Manufacturer', 'Vehicle_type']:
    car_sales_cat[columns] = car_sales_cat[columns].astype('category')
    car_sales_cat=create_dummies(car_sales_cat , columns)
    
car_sales_cat.head()
```

```python
car_sales_new = pd.concat([car_sales_num, car_sales_cat], axis=1)
car_sales_new.head(2)
```

```python
# inter-correlation 

car_sales_new.corr().head()
```

```python
#Splitting the data
feature_columns = car_sales_new.columns.difference( ['Sales_in_thousands'] )
feature_columns

```

```python
X_train, X_test, Y_train, Y_test=train_test_split(car_sales_new[feature_columns],car_sales_new['Sales_in_thousands'], test_size=0.03)
```

```python
lnrreg=LinearRegression()
lnrreg.fit(X_train, Y_train)
```

```python
y_pred = lnrreg.predict(X_train)
```

```python
print np.sqrt(metrics.mean_squared_error(Y_train,y_pred))
print metrics.r2_score(Y_train,y_pred)
```

```python
y_pred_test = lnrreg.predict(X_test)
print np.sqrt(metrics.mean_squared_error(Y_test,y_pred_test))
print metrics.r2_score(Y_test,y_pred_test)
```
